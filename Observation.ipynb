```python
import pandas as pd
from sklearn.cluster import KMeans
# from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import plotly.express as px

# Assuming 'df' is your DataFrame
df_filled = df_usage.fillna(0)  # Fill NaN values with 0

data = df_filled.iloc[:, 3:]  # Extracting only the columns with answers

# Standardize the data
scaler = StandardScaler()
# scaler = MinMaxScaler()
data_standardized = scaler.fit_transform(data)

# Choose the number of clusters (you may need to adjust this based on your data)
num_clusters = 4

# Apply k-means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=10)
df_filled['cluster'] = kmeans.fit_predict(data_standardized)

# Visualize the clusters using PCA (for 3D visualization)
pca = PCA(n_components=3)
data_pca = pca.fit_transform(data_standardized)

# Create an interactive 3D scatter plot using Plotly Express
fig = px.scatter_3d(
    df_filled,
    x=data_pca[:, 0],
    y=data_pca[:, 1],
    z=data_pca[:, 2],
    color='cluster',
    labels={'color': 'Cluster'},
    title='Clustering of Respondents (Interactive 3D)',
)

fig.show()

```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


data = df_filled.iloc[:, 3:]

# Appliquer la mÃ©thode du coude pour trouver le nombre optimal de clusters
sse = []
for k in range(1, 11):  # Testons de 1 Ã  10 clusters
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(data)
    sse.append(kmeans.inertia_)  # Somme des carrÃ©s des distances des Ã©chantillons Ã  leur centre de cluster le plus proche

# Tracer la courbe SSE par rapport au nombre de clusters
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), sse, 'bx-')
plt.xlabel('Nombre de clusters')
plt.ylabel('SSE (Somme des carrÃ©s des erreurs)')
plt.title('La MÃ©thode du Coude pour dÃ©terminer le nombre optimal de clusters')
plt.show()
```
